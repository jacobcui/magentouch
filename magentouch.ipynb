```python
from magentouch import Urls, Magentouch
from settings import username, password

urls = Urls()
urls.magento = 'http://shop.digirocks.com.au/api/?wsdl'  # soap 1.0

mgt = Magentouch(url=urls.magento)
client = mgt.get_client()

mgt.login(username, password)
```

```python
res_info = mgt.call('magento.info')
res_tree = mgt.call('catalog_category.tree')
```

```python

def translate_res(soap_item):
    """
    item = dict(key = value)
    value could be string or another list
    """
    if hasattr(soap_item, 'item'):
        return translate_res(soap_item.item)

    if isinstance(soap_item, list):
        return [translate_res(n) for n in soap_item]

    if not hasattr(soap_item, 'value'):
        return soap_item
    
    if isinstance(soap_item['value'], list):
        if isinstance(soap_item['key'], list):
            key = soap_item['key'][0]
        else:
            key = soap_item['key']
        
        return {key: [translate_res(n) for n in soap_item['value']]}

    if hasattr(soap_item, 'key'):
        return {soap_item['key']: soap_item['value']}

```
